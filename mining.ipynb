## Calculate daily views (每天平均觀看數量/總天數)

```python
udn_date = Udn['publishedAt'].value_counts().to_dict()
set_date = SETNews['publishedAt'].value_counts().to_dict()
tvbs_date = TVBS['publishedAt'].value_counts().to_dict()

def calculate_daily_view(news, date):
    daily_mean_view = []
    for k, v in date.items():
        date_video_count = news.loc[news['publishedAt'] == k]['viewCount']
        # one day video
        daily_mean_view.append(date_video_count.mean())
    return daily_mean_view
        
udn_average = calculate_daily_view(Udn, udn_date)
set_average = calculate_daily_view(SETNews, set_date)
tvbs_averag = calculate_daily_view(TVBS, tvbs_date)
```

```python
print('udn average daily video has: %d, average daily view has %f' % (len(udn_average), sum(udn_average)/len(udn_average)))
print('set average daily video has: %d, average daily view has %f' % (len(set_average), sum(set_average)/len(set_average)))
print('tvbs average daily video has: %d, average daily view has %f' % (len(tvbs_averag), sum(tvbs_averag)/len(tvbs_averag)))
```
